# 基于RFM模型的用户价值的分析
- 项目背景
A公司是一家礼品销售公司，主要经营的业务是在线礼品销售，随着近年来电商不断发展，行业竞争愈发激烈，为了保留现有用户，所以采用RFM分析模型对不同层次的用户进行划分，并对不同层次用户提供针对性的优惠政策或是服务。
-RFM
-R：Recency，用户上一次购买产品时间至今为止之差，距今越短则用户越有价值。
-F：frequency，用户消费频率
-monetary，货币，指用户在一定时间内的消费金额。

-数据描述
InvoiceNo：订单号，如果订单是C字母开头则为退货订单。
StockCode：商品编号。
Description：商品描述
Quantity：购买商品数量
InvoiceDate：订单日期
UnitPrice：商品单价
CustomerID：用户ID
Country：用户所在国家

```python
#计算消费金额
df["monetory"] = df["Quantity"] * df["UnitPrice"]
df
```

```python
#将订单日期转换为日期类型
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
```

# 建立RFM

```python
#建立R
def build_r(group_data, current):
    #去除退货订单
    t = group_data[~group_data["InvoiceNo"].str.startswith("C")]
    if len(t) == 0:
        #如果没有购买记录则将所有订单最早的日期作为最后一次购买日期
        delta = current - df["InvoiceDate"].min()
    else:
        #计算当前日期与最后一次购买的间距
        delta = current - t["InvoiceDate"].max()
    return delta.days
rfm = pd.DataFrame()
current = pd.to_datetime("2012-01-01 00:00:00")
g = df.groupby("CustomerID")
rfm["recency"] = g.apply(build_r, current)
rfm.head()
```

```python
#建立F
def build_f(group_data):
    #确认订单类型
    is_refound = group_data["InvoiceNo"].str.startswith("C")
    normal = group_data[~is_refound]
    refound = group_data[is_refound]
    num = normal["InvoiceNo"].unique().shape[0] - refound["InvoiceNo"].unique().shape[0]
    return num if num > 0 else 0
rfm["frequent"] = g.apply(build_f)
rfm
```

```python
#建立M
def build_m(group_data):
    money = group_data["monetory"].sum()
    return money if money >0 else 0
rfm["monetory"] = g.apply(build_m)
rfm
```

## 用户评级

```python
#R值越高价值越低，将R值转为负数处理
rfm["recency"] *= -1
```

```python
r_bar = rfm["recency"].mean()
f_bar = rfm["frequent"].mean()
m_bar = rfm["monetory"].mean()
```

```python
def level_evaluate(c):
    if c["monetory"] >= m_bar and c["recency"] >= r_bar and c["frequent"] >= f_bar:
        s = "重要价值客户"
    elif c["monetory"] >= m_bar and c["recency"] >= r_bar and c["frequent"] < f_bar:
        s = "重要发展客户"
    elif c["monetory"] >= m_bar and c["recency"] < r_bar and c["frequent"] >= f_bar:
        s = "重要保持客户"
    elif c["monetory"] >= m_bar and c["recency"] < r_bar and c["frequent"] < f_bar:
        s = "重要挽留客户"
    elif c["monetory"] < m_bar and c["recency"] >= r_bar and c["frequent"] >= f_bar:
        s = "一般价值客户"
    elif c["monetory"] < m_bar and c["recency"] >= r_bar and c["frequent"] < f_bar:
        s = "一般发展客户"
    elif c["monetory"] < m_bar and c["recency"] < r_bar and c["frequent"] >= f_bar:
        s = "一般保持客户"
    elif c["monetory"] < m_bar and c["recency"] < r_bar and c["frequent"] < f_bar:
        s = "一般挽留客户"
    else:
        print(c)
    return s
rfm["type"] = rfm.apply(level_evaluate, axis=1)
```

# 总结
通过建立rfm模型对该电商公司顾客群体进行分类，从而针对不同的客户群体提供合适的服务，来达到保留现有客户和挽回潜在流失顾客。在本项目中，针对rfm维度的划分使用的是平均值。也可以使用其他方法进行维度的划分，比如中位数和打分机制来划分维度。针对不同类型的公司使用合适的方法从而达到更加理想的效果。
